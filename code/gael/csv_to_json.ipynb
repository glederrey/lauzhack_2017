```python
import pandas as pd
import numpy as np

from datetime import datetime, date
import matplotlib.pyplot as plt

import json

data = 'data/'
```

```python
atms = pd.read_csv(data + 'atms.small.csv')
clients = pd.read_csv(data + 'clients.small.csv')
companies = pd.read_csv(data + 'companies.small.csv')
```

```python
nodes = {}

cols = list(atms.columns)
cols.remove('id')

for i in atms.index:
    row = atms.loc[i]
    
    tmp = {'type': 'atm'}
    for c in cols:
        tmp[c] = row[c]
    
    nodes[row['id']] = tmp
    
cols = list(companies.columns)
cols.remove('id')

for i in companies.index:
    row = companies.loc[i]
    
    tmp = {'type': 'company'}
    for c in cols:
        tmp[c] = row[c]
    
    nodes[row['id']] = tmp
    
cols = list(clients.columns)
cols.remove('id')

for i in clients.index:
    row = clients.loc[i]
    
    tmp = {'type': 'client'}
    for c in cols:
        if c == 'age':
            tmp[c] = int(row[c])
        else:
            tmp[c] = row[c]
    
    nodes[row['id']] = tmp
```
